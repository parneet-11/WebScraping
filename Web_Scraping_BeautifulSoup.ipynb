```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.geeksforgeeks.org/python/python-programming-language-tutorial/")
```

```python
print(web)
```

```python
web.content
```

```python
soup = BeautifulSoup(web.content, "html.parser")
```

```python
print(soup.prettify())
```

**Get All Text**

```python
paras = soup.find_all("p")
```

```python
print(paras)
```

```python
for l in paras:
  print(l.text)
```

**Extract Link**

```python
links = soup.find_all("a")
```

```python
links
```

```python
for i in links:
  print(i.get("href"))
```

**Extract All Images**

```python
images = soup.find_all("img")
images
```

```python
for i in images:
  print(i.get("src"))
```
